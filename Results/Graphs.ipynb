```python
import matplotlib.pyplot as plt
import numpy as np
```

# Marginal Release AccidentalDrugDeaths

```python
x = [0.1, 0.2, 0.5, 1, 2, 5, 10]
y = [0.16540942650588672, 0.12081720312970494, 0.08165848231349307, 0.06527973712192091, 0.06132411128369232, 0.04899592315808463, 0.03926967193959008]
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel('epsilon')
plt.ylabel('average l1 error')
plt.title('Marginal Release Score (US Accidental Drug Deaths)')
plt.savefig('Plot1.png')
```
